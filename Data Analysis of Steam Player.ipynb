**Top 10 Game of 2021**

```python
df2 = df
df2['Date'] = pd.to_datetime(df2['Date'])
df2_2021 = df2[df2['Date'].dt.year == 2021]
top10_peak_2021 = df2_2021[['Game_Name', 'Peak_Players']].groupby('Game_Name') \
                                                                    .sum() \
                                                                    .sort_values('Peak_Players', ascending=False) \
                                                                    .head(10)
top10_avg_2021 = df2_2021[['Game_Name', 'Avg_players']].groupby('Game_Name') \
                                                                    .sum() \
                                                                    .sort_values('Avg_players', ascending=False) \
                                                                    .head(10)
fig, axes = plt.subplots(2, figsize=(10, 12))
fig.suptitle('Top 10 Game in 2021')
sns.barplot(ax=axes[0], data=top10_peak_2021, x='Peak_Players', y=top10_peak_2021.index)
axes[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000) + 'K'))
axes[0].set(title='Highest peak players', xlabel='Peak Players', ylabel='Game Name')
sns.barplot(ax=axes[1], data=top10_avg_2021, x='Avg_players', y=top10_avg_2021.index)
axes[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000) + 'K'))
axes[1].set(title='Highest average players', xlabel='Average Players', ylabel='Game Name')
```

We can observe that games rank 1 to 9 for both 'Average' and 'Peak' players are the same. There is only one difference in the last ranking.

**Average Player by Months**

```python
top10 = df2[df2['Game_Name'].isin(top10_avg_2021.index)]
top10
```

```python
top10 =top10.sort_values(by="Date")
top10
```

```python
fig, axes = plt.subplots(2, figsize=(20, 18))
fig.suptitle('Average Player By Month')
sns.lineplot(ax=axes[0], data=top10, x='Date', y='Avg_players', hue='Game_Name')
axes[0].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000) + 'K'))
axes[0].tick_params(axis='x', rotation=90)
axes[0].set(ylabel='Avg Player')
sns.lineplot(ax=axes[1], data=top10, x='Date', y='Avg_players', hue='Game_Name')
axes[1].tick_params(axis='x', rotation=90)
axes[1].set(ylabel='Avg Player (log)', yscale='log')
```

Between 2017 and 2018, the average number of PUBG players climbed dramatically, according to the graph. The average PUBG player began to fall after 2018. The typical Dota2 and CSGO gamer, on the other hand, is steadily increasing. The average Dota2 and CSGO player has remained consistent since 2015. Keeping everyone at home in the Covid-19 scenario in early 2021Â greatly boosted the number of players.
